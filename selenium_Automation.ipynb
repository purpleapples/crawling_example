```python
from selenium import webdriver
# chrome browser 실행
driver = webdriver.Chrome("D:/data_science/driver/chromedriver.exe") # 드라이버 실행
# ww.naver.com에 접속
driver.get("http://www.naver.com")
# 잠시만 대기
driver.implicitly_wait(3)

# 로그인 버튼의 xpath : //*[@id="account"]/a
btn_login = driver.find_elements_by_xpath("//*[@id=\"account\"]/a")[0]
btn_login.click() # 클릭 이벤트 전송

# 잠시만 대기
driver.implicitly_wait(3)

# id 입력창을 찾아서 키보드 입력을 전송
# 비밀번호 입력창을 찾아서 키보드 입력을 전송
# 로그인 버튼을 누른다.

input_id = driver.find_element_by_css_selector("#id")
input_id.send_keys("yourid")
# 비밀번호 입력창을 찾아서 키보드로 전송

password = driver.find_element_by_css_selector("#pw")
password.send_keys("yourpassword")

```

```python
# 네이버에 접속
driver = webdriver.Chrome("D:/data_science/driver/chromedriver.exe")
driver.get("http://www.naver.com")
driver.implicitly_wait(3)
# 검색어 창에 검색어를 입력
input_search = driver.find_element_by_xpath("//*[@id=\"query\"]")
input_search.send_keys("갤럭시 FE")
# 검색 버튼 눌러
btn_submit = driver.find_element_by_css_selector("#search_btn")
btn_submit.click()
# 페이지 소스 확보

print("검색 페이지의 소스")
html_text = driver.page_source

from bs4 import BeautifulSoup
dom = BeautifulSoup(html_text)

print(dom.prettify())
```
